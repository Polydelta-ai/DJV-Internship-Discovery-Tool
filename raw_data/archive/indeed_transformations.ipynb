```python
s = 'This is a test sentence that has requirements and look for: in it'
word = 'requirements'

# Mark requirements
# Mark responsibilities
# split the sections
# assign the sections to new pandas series
# make sure you have a default value

# return both sections -  df[a], df[b] = return sectiona, sectionb


def get_responsibility_section(job_description): #, responsibility_breaks, qualification_breaks)
    responsibility_breaks =[
        "duties include:",
        "what you\'ll do:",
        "what you will do:"
        "key accountabilities",
        "primary role",
        "role:",
        "achieve:",
        "responsibilities",
        "responsibility",
    ]

    qualification_breaks = [
        "requirements",
        "qualifications",
        "bring to the team:",
        "make a strong candidate:",
        "looking for:",
        "look for:",
    ]
    
    
    for word in responsibility_breaks:
        if word in job_description.lower():
            print('found word')
            job_description  = job_description.replace(word, "***responsibilities---")
            break
    
    for word in qualification_breaks:
        if word in job_description.lower():
            job_description  = job_description.replace(word, "***qualifications---")
            break
    
    responsibility_section = ''

    job_description_sections = job_description.split("***")

    for section in job_description_sections:
        if "responsibilities---" in section:
            responsibility_section = section.replace("---","")
        if "qualifications---" in section:
            qualification_section = section.replace('---','')

    return responsibility_section
    

def get_qualification_section(job_description): #, responsibility_breaks, qualification_breaks)
    responsibility_breaks =[
        "duties include:",
        "what you\'ll do:",
        "what you will do:"
        "key accountabilities",
        "primary role",
        "role:",
        "achieve:",
        "Responsibilities",
        "Responsibility",
    ]

    qualification_breaks = [
        "requirements",
        "qualifications",
        "bring to the team:",
        "make a strong candidate:",
        "looking for:",
        "look for:",
    ]


    for word in responsibility_breaks:
        if word in job_description:
            job_description  = job_description.replace(word, "***responsibilities---")
            break

    for word in qualification_breaks:
        if word in job_description.lower():
            job_description  = job_description.replace(word, "***qualifications---")
            break

    qualification_section = ''

    job_description_sections = job_description.split("***")
    for section in job_description_sections:
        if "responsibilities---" in section:
            responsibility_section = section.replace("---","")
        if "qualifications---" in section:
            qualification_section = section.replace('---','')

    return qualification_section

```

```python
indeed['responsibilities_section'] = indeed.common_description.apply(lambda x: get_responsibility_section(x))
indeed['qualifications_section'] = indeed.common_description.apply(lambda x: get_qualification_section(x))
```
